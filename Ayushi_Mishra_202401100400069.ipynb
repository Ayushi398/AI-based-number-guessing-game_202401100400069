```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def ai_guess_number(low=1, high=100):
    target = np.random.randint(low, high + 1)  # Random number to guess
    attempts = []
    guess = None
    left, right = low, high

    while guess != target:
        guess = (left + right) // 2  # Binary search strategy
        attempts.append(guess)
        if guess < target:
            left = guess + 1
        elif guess > target:
            right = guess - 1

    return target, attempts

def plot_attempts(attempts, target):
    df = pd.DataFrame({'Attempt': range(1, len(attempts) + 1), 'Guess': attempts})

    sns.set_style("darkgrid")
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Line Plot
    sns.lineplot(data=df, x='Attempt', y='Guess', marker='o', ax=axes[0, 0])
    axes[0, 0].axhline(target, color='r', linestyle='--')
    axes[0, 0].set_title('Line Plot')

    # Scatter Plot
    sns.scatterplot(data=df, x='Attempt', y='Guess', ax=axes[0, 1])
    axes[0, 1].axhline(target, color='r', linestyle='--')
    axes[0, 1].set_title('Scatter Plot')

    # Bar Plot
    sns.barplot(data=df, x='Attempt', y='Guess', ax=axes[1, 0])
    axes[1, 0].axhline(target, color='r', linestyle='--')
    axes[1, 0].set_title('Bar Plot')

    # Histogram
    sns.histplot(df['Guess'], bins=len(df), kde=True, ax=axes[1, 1])
    axes[1, 1].axvline(target, color='r', linestyle='--')
    axes[1, 1].set_title('Histogram')

    plt.tight_layout()
    plt.show()

# Run the AI Number Guessing Game
target, attempts = ai_guess_number()
print(f"Target Number: {target}, Attempts: {len(attempts)}")
plot_attempts(attempts, target)
```
